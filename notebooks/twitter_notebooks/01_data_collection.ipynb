# Twitter Data Collection
**Goal:** Scrape tweets about Wikipedia, AI's impact on knowledge, and related topics using `snscrape` (no API key required). Mirrors the structure of `reddit_notebooks/01_data_collection.ipynb`.

**Output:** `data/raw/twitter_raw/twitter_raw.csv`

```python
import sys
import os
sys.path.insert(0, os.path.abspath('../..'))

import pandas as pd
from src.twitter_scraper import TwitterScraper

print('Imports OK')
```

## 1. Define Search Queries
We define 7 queries that cover:
- Core Wikipedia brand
- AI/ChatGPT vs Wikipedia
- Wikipedia reliability/criticism
- Wikipedia vs other knowledge sources
- Community & editor discussion

```python
QUERIES = [
    "wikipedia",
    "wikipedia ChatGPT OR AI",
    "wikipedia reliable OR unreliable OR wrong OR accurate",
    "wikipedia OR wikimedia editors OR editing",
    "wikipedia vs google OR ChatGPT",
    "donate wikipedia OR wikimedia",
    "wikipedia misinformation OR vandalism OR bias",
]

LIMIT_PER_QUERY = 500   # adjust up/down depending on time constraints
SINCE           = '2020-01-01'

print(f'Total queries: {len(QUERIES)}')
print(f'Max tweets per query: {LIMIT_PER_QUERY}')
print(f'Date filter: >= {SINCE}')
```

## 2. Run Scraper

```python
scraper = TwitterScraper()
df_raw  = scraper.scrape_multiple_queries(
    queries          = QUERIES,
    limit_per_query  = LIMIT_PER_QUERY,
    since            = SINCE,
)

print(f'\nRaw shape: {df_raw.shape}')
df_raw.head(3)
```

## 3. Quick Sanity Checks

```python
print('Dtypes:')
print(df_raw.dtypes)
print()
print('Null counts:')
print(df_raw.isnull().sum())
print()
print('Tweet count per query:')
print(df_raw['query'].value_counts())
```

```python
print('Date range:')
print(f"  earliest: {df_raw['date'].min()}")
print(f"  latest:   {df_raw['date'].max()}")
print()
print('Engagement summary:')
df_raw[['likes','retweets','replies']].describe()
```

## 4. Save Raw Data

```python
OUT_DIR = os.path.abspath('../../data/raw/twitter_raw')
os.makedirs(OUT_DIR, exist_ok=True)

out_path = os.path.join(OUT_DIR, 'twitter_raw.csv')
df_raw.to_csv(out_path, index=False)
print(f'Saved {len(df_raw)} tweets → {out_path}')
```
