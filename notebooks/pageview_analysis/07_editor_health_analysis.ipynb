# Editor Health Analysis

This notebook analyzes the trends in Wikipedia editor participation, specifically the number of active editors and their edit volume over time.

```python
import sys
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Add src to path
sys.path.append(os.path.abspath('../../src'))
from api_client import WikipediaAPIClient
```

```python
client = WikipediaAPIClient()
project = "en.wikipedia.org"
start_date = "20100101"
end_date = "20240101"

print("Fetching editor metrics...")
editors_df = client.get_editor_metrics(project, start_date, end_date)
editors_df.head()
```

```python
# Plotting the number of active editors
plt.figure(figsize=(14, 7))
plt.plot(editors_df['timestamp'], editors_df['net_new'], label='Net New Editors', alpha=0.5)
# Note: The exact column name might vary depending on API response details. 
# Let's assume 'results' contains the counts if 'net_new' isn't what we want.
plt.title(f'Wikipedia Active Editor Trends ({project})', fontsize=16)
plt.ylabel('Number of Editors')
plt.grid(True)
plt.show()
```
