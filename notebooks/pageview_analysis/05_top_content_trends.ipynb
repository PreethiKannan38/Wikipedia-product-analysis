# Top Content Trends Analysis

This notebook explores the most viewed articles on Wikipedia over time to identify trending topics and content patterns.

```python
import sys
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Add src to path
sys.path.append(os.path.abspath('../../src'))
from api_client import WikipediaAPIClient
```

```python
client = WikipediaAPIClient()
project = "en.wikipedia.org"
year = "2023"
month = "10"

print(f"Fetching top articles for {year}-{month}...")
top_articles = client.get_top_articles(project, year, month)
top_articles.head(20)
```

```python
# Filter out main page and search pages if necessary
excluded = ['Main_Page', 'Special:Search']
filtered_top = top_articles[~top_articles['article'].isin(excluded)].head(15)

plt.figure(figsize=(12, 8))
sns.barplot(data=filtered_top, x='views', y='article', palette='viridis')
plt.title(f'Top Trending Wikipedia Articles ({year}-{month})', fontsize=16)
plt.xlabel('Pageviews', fontsize=12)
plt.ylabel('Article Title', fontsize=12)
plt.show()
```

### Conclusion
The analysis reveals the dominant interests of users during this period, often driven by current events, media releases, or seasonal trends.
